### Results for the Entanglement-based Quantum Key Distribution in a $n$ x $n$ grid topology using Netsquid

```python
import matplotlib.pyplot as plt
import pandas

res = pandas.read_csv("results.csv")

fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(res["path_length"], res["fidelity_rate"], 'bo')
axs[0].set_xlabel('Path length')
axs[0].set_ylabel('Fidelity rate')
axs[0].set_title("Path length v/s Fidelity rate")

axs[1].plot(res["path_length"], res["loss"], 'ro')
axs[1].set_xlabel('Path length')
axs[1].set_ylabel('Qubit Loss')
axs[1].set_title("Path length v/s Qubit Loss")

plt.show()
```
