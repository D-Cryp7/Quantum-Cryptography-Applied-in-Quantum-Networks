```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

results = pd.read_csv("results.csv", index_col = 0)
results
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
for col in results.columns:
    # col = results.columns[0] # 10
    
    col_data = results[col]
    data = pd.DataFrame()
    for i in col_data.index:
        if not pd.isnull(col_data[i]):
            temp = col_data[i].replace("nan", "0.0")
            temp = eval(temp)
            data[i] = temp
    if not data.empty:
        data = data.agg(['mean', 'sem']).T.rename(columns = {'mean': 'fidelity'})
        data.plot(y = 'fidelity', yerr = 'sem', label = f"{col} km", ax = ax)

plt.legend(loc = 1, prop = {'size': 20})
plt.xlabel("path length (number of nodes)", fontsize = 20)
plt.ylabel("fidelity", fontsize = 20)
plt.title("Path length v/s Mean fidelity in a 10x10 Grid Quantum Network", fontsize = 20)
plt.xticks(range(2, i), fontsize = 18)
plt.yticks(fontsize = 18)
plt.xlim([1.5, 8])
plt.show()
```
