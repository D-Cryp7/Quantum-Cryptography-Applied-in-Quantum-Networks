**Attemps until successful entanglement (fidelity = 1)**

```python
n = 3
length = 3
results_df = pandas.read_csv(f"Results/entanglement_probability_results_{n}x{n}_length_{length}.csv")
fidelity_results = results_df["fidelity"].tolist()
runs = zero_runs(fidelity_results)
data = runs[:,1] - runs[:,0]

plt.figure(figsize=(15,15))
plt.xlim([min(data) - 1, max(data)])
plt.hist(data, bins = 90, alpha = 0.5)
plt.title(f'Attempts to achieve a fidelity of entanglement equal to 1 in a path of length {length}')
plt.xlabel('Attemps')

# plt.savefig(f"AUSE_{n}x{n}_length_{length}.svg")

plt.show()
```

**Consecutive successful entanglements**

```python
n = 3
length = 5
results_df = pandas.read_csv(f"Results/entanglement_probability_results_{n}x{n}_length_{length}.csv")
fidelity_results = results_df["fidelity"].tolist()
fidelity_results = [int(i) ^ 1 for i in fidelity_results]
runs = zero_runs(fidelity_results)
data = runs[:,1] - runs[:,0] - 1

plt.figure(figsize=(15,15))
plt.xlim([min(data) + 1, max(data)])
plt.ylim([0, 16000])
plt.hist(data, bins = 90, alpha = 0.5)
plt.title(f'Consecutive successful entanglements in a path of length {length}')
plt.xlabel('Consecutive successful entanglements')

plt.savefig(f"CSE_{n}x{n}_length_{length}.svg")

plt.show()
```
